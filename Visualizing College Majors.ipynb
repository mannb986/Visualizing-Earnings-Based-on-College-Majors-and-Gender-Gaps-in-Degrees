# Job Outcomes of Students Who Graduated Between 2010 & 2012

The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released iton the Github repo.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

```python
raw_data_count = len(recent_grads)
print(raw_data_count)
```

Dropping rows with missing values, so I don't generate any errors using Matplotlib. 

```python
recent_grads = recent_grads.dropna(axis=0)
```

```python
cleaned_data_count = len(recent_grads)
print(cleaned_data_count)
```

## Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women')
```

### Findings

The data suggests that students in less popular majors had some of the hihest median salaries. 

The data suggests that the majority of students graduating into full time employment have a median salary of under $40k.
